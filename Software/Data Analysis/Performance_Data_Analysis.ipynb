# Performance Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Path Set-Up
training_run_folder = "train" 

# Ruta al archivo results.csv
base_path = "c:/Users/usuario/Documents/Luis Adrian Cabrera/Proyectos/Representative_teams/Fox Robotics/Cubesat/CubeRT_Cubesat/Software/Machine_Learning_Solutions/runs/detect"
results_file_path = os.path.join(base_path, training_run_folder, "results.csv")

# Cargar los datos del archivo CSV
try:
    results_df = pd.read_csv(results_file_path)
except FileNotFoundError:
    print(f"Error: El archivo '{results_file_path}' no fue encontrado.")
    print("Por favor, verifica la ruta y el nombre de la carpeta de entrenamiento.")
    # Salir si el archivo no se encuentra para evitar más errores
    raise SystemExit()

# Limpiar los nombres de las columnas
results_df.columns = results_df.columns.str.strip()

# # Observar las columnas disponibles 
# print("Columnas disponibles en el CSV:")
# print(results_df.columns)
# print("\nPrimeras filas del DataFrame:")
# print(results_df.head())

# Variables para columnas
epoch_col = 'epoch'
precision_col = 'metrics/precision(B)'
recall_col = 'metrics/recall(B)'
map50_col = 'metrics/mAP50(B)'
map50_95_col = 'metrics/mAP50-95(B)'

# Imprimir las últimas métricas
if not results_df.empty:
    last_epoch_data = results_df.iloc[-1]
    print("--- Métricas Finales (Última Época) ---")
    print(f"Run: {training_run_folder}")
    print(f"Época: {last_epoch_data[epoch_col]}")
    print(f"Precisión (Precision):     {last_epoch_data[precision_col]:.4f}")
    print(f"Recall:                  {last_epoch_data[recall_col]:.4f}")
    print(f"mAP@0.50:                {last_epoch_data[map50_col]:.4f}")
    print(f"mAP@0.50-0.95:           {last_epoch_data[map50_95_col]:.4f}")
    print("--------------------------------------\n")
else:
    print("El DataFrame de resultados está vacío, no se pueden mostrar las métricas finales.")


# Generar gráficas
plt.style.use('seaborn-v0_8-whitegrid') # Estilo de gráfica

# Gráfica de Precisión
plt.figure(figsize=(10, 6))
plt.plot(results_df[epoch_col], results_df[precision_col], marker='o', linestyle='-', color='b')
plt.title(f'Evolución de la Precisión (Precision) - Run: {training_run_folder}')
plt.xlabel('Época')
plt.ylabel('Precisión')
plt.ylim(0, 1.0)
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfica de Recall
plt.figure(figsize=(10, 6))
plt.plot(results_df[epoch_col], results_df[recall_col], marker='o', linestyle='-', color='r')
plt.title(f'Evolución del Recall - Run: {training_run_folder}')
plt.xlabel('Época')
plt.ylabel('Recall')
plt.ylim(0, 1.0)
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfica de mAP@0.50
plt.figure(figsize=(10, 6))
plt.plot(results_df[epoch_col], results_df[map50_col], marker='o', linestyle='-', color='g')
plt.title(f'Evolución de mAP@0.50 - Run: {training_run_folder}')
plt.xlabel('Época')
plt.ylabel('mAP@0.50')
plt.ylim(0, 1.0)
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfica de mAP@0.50-0.95
plt.figure(figsize=(10, 6))
plt.plot(results_df[epoch_col], results_df[map50_95_col], marker='o', linestyle='-', color='purple')
plt.title(f'Evolución de mAP@0.50-0.95 - Run: {training_run_folder}')
plt.xlabel('Época')
plt.ylabel('mAP@0.50-0.95')
plt.ylim(0, 1.0)
plt.grid(True)
plt.tight_layout()
plt.show()
```
